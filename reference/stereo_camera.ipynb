```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
```

```python
# cap.set(cv2.CAP_PROP_ZOOM, 200)
```

```python
if not cap.isOpened():
    print("Camera not found")
    exit()

cap.set(cv2.CAP_PROP_FOURCC, cv2.VideoWriter_fourcc(*'MJPG'))
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
print(f"{int(width)}x{int(height)}")
print(cap.get(cv2.CAP_PROP_ZOOM))
```

```python
while True:
    ret, frame = cap.read()

    if not ret:
        print("Do not read frame")
        break

    h, w  = frame.shape[:2]

    # 가로 분할
    half_width = w // 2

    left_camera = frame[0:h, 0:half_width]
    right_camera = frame[0:h, half_width:w]

    # 필터
    # left_camera_gray = cv2.cvtColor(left_camera, cv2.COLOR_BGR2GRAY)
    # left_camera = cv2.cvtColor(left_camera_gray, cv2.COLOR_GRAY2BGR)
    # 그레이스케일 변환
    left_gray = cv2.cvtColor(left_camera, cv2.COLOR_BGR2GRAY)
    right_gray = cv2.cvtColor(right_camera, cv2.COLOR_BGR2GRAY)

    # 합치기
    combined_frame = cv2.hconcat([right_camera, left_camera])

    cv2.imshow("View", combined_frame)
    
    if cv2.waitKey(1) & 0xff == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
